# 離散化の考え方
## ボールの運動問題
- 時間経過とともに動くボールの軌跡をある時間間隔ごとに離散化してプロットする
- ここでは、時間刻み`dt`が離散点(ステンシル)となっており、これを多く取ることで解析解と近い精度を持つ数値解が得られる

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

g = 9.8 # 重力加速度

v0 = 10 # 初速度
h0 = 0 # 初期高さ
dt = 0.02 # 時刻刻み

plt.figure(figsize=(7, 7), dpi = 100) # グラフのサイズ
plt.rcParams["font.size"] = 25 # グラフの文字サイズ

# 解析解(Analiytical Solution)
t = np.linspace(0, 2 * v0 / g, 100)
h = -0.5 * g * t ** 2 + v0 * t + h0
la, = plt.plot(t, h, color='blue')

# 数値解(Numerical Solution)
t = 0
h = h0
while h >= h0:
    ln = plt.scatter(t, h, marker='o', c='black')
    h += (-g * t + v0) * dt
    t += dt

# グラフの後処理
plt.grid(color='black', linestyle='dashed', linewidth=0.5)
plt.xlabel('Time')
plt.ylabel('Height')
plt.legend(handles = [la, ln], labels = ['Analytical', 'Numerical'])
plt.show()
```
